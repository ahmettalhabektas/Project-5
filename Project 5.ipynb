# Spliting data

```python
X=train.drop(41,axis=1)
y=train.iloc[:,-1]
Z=test
```

```python
from sklearn.naive_bayes import GaussianNB , BernoulliNB # 2 ÅŸey iÃ§in
from sklearn.naive_bayes import MultinomialNB# 3 ve daha fazla iÃ§in
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
g=GaussianNB()
b=BernoulliNB()
k=KNeighborsClassifier(n_neighbors=10)
d=DecisionTreeClassifier()
gbc=GradientBoostingClassifier()
log=LogisticRegression()
svc=SVC()
```

```python
#Ã¶nce yukarda olduÄŸu gibi tanÄ±mlamak lazÄ±m
algorithms = [g,b,k,log,d,gbc]
names = ['GaussianNB', 'BernoulliNB', 'K Nearest', 'Logistic', 'Single Tree', 'Gradient Boost']
```

```python
for i in range(len(algorithms)):
    algorithms[i] = algorithms[i].fit(X, y)
# print metrics 
for i in range(len(algorithms)):
    pred=algorithms[i].predict(Z)
    submission = pd.DataFrame(pred)
    submission.columns = ['Solution']
    submission['Id'] = np.arange(1,submission.shape[0]+1)
    submission = submission[['Id', 'Solution']]
    submission.to_csv('submission_{}.csv'.format(algorithms[i]), index=False)
```
